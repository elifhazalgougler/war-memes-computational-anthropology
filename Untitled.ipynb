```python
%matplotlib inline

import seaborn as sns
import matplotlib.pyplot as plt

ct = pd.crosstab(df["main_theme"], df["main_meme_theme"])

plt.figure(figsize=(10,6))

sns.heatmap(
    ct,
    annot=True,
    fmt="d",
    cmap="Blues",
    linewidths=0.5
)

plt.title("Relationship Between Caption Themes and Meme Themes")
plt.xlabel("Meme Theme")
plt.ylabel("Caption Theme")

plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```
